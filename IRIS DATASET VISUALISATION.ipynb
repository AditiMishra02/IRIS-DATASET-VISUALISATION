```python
# Importing required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
# Reading the data
df = pd.read_csv("Iris.csv")
```

```python
# printing the data
print(df.head(10))  # prints the first 10 rows
print(df.tail(10))  # prints the last 10 rows
print(df.shape)  # prints the number of rows and columns
```

```python
# Determining the feature having null values and replacing them with zero
for col in df.columns:

    df[col] = df[col].replace(np.nan, 0)  # replacing the NAN with 0
    print(col, df[col].isnull().sum())   # to check the null values
```

```python
# striplot is used to plot the graph
sns.stripplot(x="Species", y="PetalLengthCm", data=df)
```
